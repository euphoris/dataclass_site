## 템플릿 매칭

```python
img = cv2.imread('figures.jpg')  # 큰 이미지
template = cv2.imread('taekwonv1.jpg')  # 템플릿
template = template[:, 50:-80] # 여백 잘라냄
th, tw, _ = template.shape  # 템플릿 크기
res = cv2.matchTemplate(img, template, cv2.TM_SQDIFF) # 템플릿 매칭
```

## 코너 검출

```python
img = cv2.imread('house.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
corners = cv2.cornerHarris(gray, 2, 3, 0.04)  # 해리스 코너 검출
coord = np.where(corners > 0.1* corners.max()) # 최댓값 10% 이상인 좌표 구하기
coord = np.stack((coord[1], coord[0]), axis=-1) # 좌표 순서 변경
dst = img.copy()
for x, y in coord:
    cv2.circle(dst, (x,y), 5, (0,0,255), 1, cv2.LINE_AA)
show(dst)
```

```python
corners = cv2.goodFeaturesToTrack(
    gray, # 입력 이미지: 8비트 또는 32비트의 단일 채널 이미지
    100, # 코너 최댓값: 검출할 최대 코너의 수
    0.01, # 코너 품질: 반환할 코너의 최소 품질(0.0 ~ 1.0 사이, 일반적으로 0.01 ~ 0.10 사용)
    15, # 최소 거리: 검출된 코너들의 최소 근접 거리 
    blockSize=3, # 블록 크기: 코너를 계산할 때, 고려하는 주변 영역의 크기
    useHarrisDetector=True, # 해리스 코너 검출기 유/무
    k=0.03) # 해리스 코너 계수
dst = img.copy() # image -> img
for corner in corners:
    corner = tuple(corner.astype(int).flatten())
    cv2.circle(dst, corner, 3, (0, 0, 255), 2)
show(dst)
```
