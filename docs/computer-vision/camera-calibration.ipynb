```python
#!hide
from PIL import Image
import cv2
import os
def show(image):
    rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # BGR → RGB로 채널 순서 변경
    return Image.fromarray(rgb) # 객체 변환: 넘파이 배열 → PIL 형식으로
os.chdir('data')
```

```python
import cv2
src = cv2.imread('chessboard_00.png')
gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)  # 흑백으로
patternSize = 6, 9  # + 모양 교차점의 개수(6행 9열)
flags = (cv2.CALIB_CB_ADAPTIVE_THRESH | cv2.CALIB_CB_FAST_CHECK |             
         cv2.CALIB_CB_NORMALIZE_IMAGE)
# 교차 찾기(찾는데 성공하면 retval == True)
retval, corners = cv2.findChessboardCorners(gray, patternSize, flags=flags)
```

그래서 이제 그 레트발 이렇게 하면은 트루라고 나오면 교차점을 다 찾았다 여섯 개 아홉 개를 다 6 곱하기 9개니까 54개를 다 찾았다는 얘기고 아유 코너에 그래서 좌표가 나옵니다

```python
# 11x11 범위에서 더 정확하게 찾기. (-1, -1)은 제외하는 범위가 없다는 뜻.
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001)
corners2 = cv2.cornerSubPix(gray, corners, (11,11), (-1,-1), criteria)
# 찾은 교차점을 원본 이미지에 추가
img_corner = cv2.drawChessboardCorners(src, patternSize, corners2, retval)
```
